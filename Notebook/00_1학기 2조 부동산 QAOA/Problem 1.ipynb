```python
import pandas as pd 
import numpy as np
```

```python
# Read data and clean data
data = pd.read_csv('final_data.csv', encoding = "CP949")
data = pd.DataFrame(data)
data = data.drop(['Unnamed: 0'], axis = 1)
print(data.head())
print(data.tail())
```

```python
# Use first 6 assets and split data into 70% train data and 30% test data
# 'Geyonggi, Hwaseong, Cheongye-dong' has no contract about first 3 years, so cubicspline works not properly.
data = data.iloc[:, :6]

train_data = data.iloc[:84, :]
test_data = data.iloc[84:, :]
print(train_data.head())
print(test_data.head())
```

```python
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
```

```python
# Apply cubicspline to NA imputation
for i in range(6): 
    cs = train_data.iloc[:, i].dropna()
    idx = cs.index.tolist()
    cs = cs.tolist()
    cs = CubicSpline(idx, cs, bc_type='natural')
    train_data.iloc[:, i] = cs(np.arange(84))

print(train_data)
```

```python
train_data_array=np.array(train_data)
print(train_data_array)

# return matrix
r=np.ndarray((83,6))
```

```python
# calculate return
# Applied ** 12 for each return to make the return as annual return.
for i in range(r.shape[1]): 
    r[:, i] = ((train_data_array[1:84, i] / train_data_array[0:83, i]) ** 12 ) - 1 
print(r)
```

```python
# Adjusted returns to account for current prices

price = train_data_array[train_data_array.shape[0]-1,:]
adj = price / np.mean(price)
r_adj = r * adj
print(r_adj)
```

```python
# Calculate return and covariance matrix for training data
mu = np.mean(r_adj,axis = 0)
sigma = np.cov(r_adj.T)
```

```python
from qiskit import Aer
from qiskit.algorithms import VQE, QAOA, NumPyMinimumEigensolver
from qiskit.algorithms.optimizers import COBYLA
from qiskit.circuit.library import TwoLocal
from qiskit.utils import QuantumInstance
from qiskit_optimization.algorithms import MinimumEigenOptimizer
from qiskit_optimization.applications import OptimizationApplication
from qiskit_optimization.converters import QuadraticProgramToQubo
import qiskit_optimization
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
# Set number of assets and risk appetite

num = 6
q = 0.5
```

```python
# Code to display results

def index_to_selection(i, num_assets):
    s = "{0:b}".format(i).rjust(num_assets)
    x = np.array([1 if s[i] == "1" else 0 for i in reversed(range(num_assets))])
    return x
```

```python
# Code to display results

def print_result(result, problem):
    selection = result.x
    value = result.fval
    print("Optimal: selection {}, value {:.4f}".format(selection, value))
    
    list_a = list(result.min_eigen_solver_result.eigenstate.keys())
    selection = np.ndarray(np.size(list_a))
    value = np.ndarray(np.size(list_a))
    for i in range(np.size(list_a)):
        value[i] = QuadraticProgramToQubo().convert(problem).objective.evaluate(index_to_selection(int(list_a[i],2),num))
        selection[i] = list_a[i][::-1]
    probabilities = np.square(list(result.min_eigen_solver_result.eigenstate.values()))
    
    d = {'selection': selection, 'value': value, 'prob':probabilities}
    df = pd.DataFrame(d)
    df_sorted = df.sort_values(by = 'prob',ascending = False)
    
    print("\n----------------- Full result ---------------------")
    print(df_sorted)
```

```python
# Simple quadratic problem (Problem 1, Training data)

qp = qiskit_optimization.QuadraticProgram()

for i in range(num): 
    qp.binary_var(name="x%d" % i)

qp.minimize(quadratic = q * sigma, linear = -1 * mu)
print(qp.export_as_lp_string())
```

```python
# Result of binary quadratic problem with qasm simulator and QAOA (Problem 1, Training data)

backend = Aer.get_backend("qasm_simulator")

cobyla = COBYLA()
cobyla.set_options(maxiter=1000)
quantum_instance = QuantumInstance(backend=backend)
qaoa_mes = QAOA(optimizer=cobyla, reps=10, quantum_instance=quantum_instance)
qaoa = MinimumEigenOptimizer(qaoa_mes)
result_prob1 = qaoa.solve(qp)

print(result_prob1)
```

```python
# Result for QAOA (Problem 1, Training data)

print_result(result_prob1, qp)
```

```python
# Eigenvector, eigenvalue and parameters (Problem 1, Training data)

print(result_prob1.min_eigen_solver_result)
```

```python
# Mean portfolio return for QAOA (Problem 1, Training data)

mean_portfolio_return = np.matmul(mu,result_prob1.x)/sum(result_prob1.x)
print(mean_portfolio_return)
```

```python
# Portfolio variance for QAOA (Problem 1, Training data)

portfolio_var = (result_prob1.x.T @ sigma @ result_prob1.x)/sum(result_prob1.x)
print(portfolio_var)
```

```python
# Required fund to construct portfolio and Expected profit

print("Required Funds :", result_prob1.x.T @ price)
print("Profit :", mean_portfolio_return * (result_prob1.x.T @ price))
```

```python
# Code for Grid Search

gridsearch = np.ndarray(2 ** num)
index = np.ndarray(2 ** num)

for i in range(2 ** num):
    gridsearch[i] = QuadraticProgramToQubo().convert(qp).objective.evaluate(index_to_selection(i,num))

for i in range(2 ** num):
    index[i] = gridsearch.argsort()[i]

# Results (compare with results from above)
print(format(int(index[0]),"b")[::-1])
print(sorted(gridsearch)[0])
```
