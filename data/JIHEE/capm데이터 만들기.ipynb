# 2. 데이터 형식 수정하기

분석에 용이하도록 데이터 형식을 수정하려고 한다!

```python
# 위에서 만든 데이터 불러오기
data = pd.read_csv('all_period.csv')
```

```python
# 데이터 정보부터 확인해봅시다(https://hogni.tistory.com/5)
data.info()
```

분석에 사용해야 하는 거래금액이 object 형식으로 되어 있음 => 형식 변화 필요  

하지만 head()로 관찰한 결과 ','부터 제거해야 함 => 정규 표현식 사용하기
reference: https://seong6496.tistory.com/242

```python
# '거래금액(만원)' 열에서 특수문자(',') 제거하기
data['거래금액(만원)'] = data['거래금액(만원)'].str.replace(pat=r',',repl=r'')
```

```python
# 잘 제거되었는지 확인하기 => 성공!
data.head()
```

```python
# 거래금액 형식 바꾸기!(astype) => 성공
data = data.astype({'거래금액(만원)':'int'})
data.info()
```

```python
# 계약월을 두자리수로 통일시키기 (str.zfill 사용)
data = data.astype({'계약월':'str'})
data["계약월"] = data["계약월"].str.zfill(2)
data
```

```python
# 계약년과 계약월을 결합하여 계약날짜라는 새로운 feature를 생성함
data = data.astype({'계약년':'str'})
data['계약날짜'] = data['계약년'] +  data['계약월']
data
```

```python
# # 의미가 없어진 계약년, 계약월 자료 없애기
# data = data.drop(['계약년','계약월'],axis=1)
```

```python
# 돟이라는 열 없애기
data = data.drop(['돟'],axis=1)
```

```python
data
```

```python
# 분기 만들기
def divide_quart(x):
  if x in ["01", "02", "03"]:
    return 1
  elif x in ["04", "05", "06"]:
    return 2
  elif x in ["07", "08", "09"]:
    return 3
  elif x in ["10", "11", "12"]:
    return 4
```

```python
data['분기'] = data['계약월'].apply(divide_quart)
```

```python
# 면적기준 만들기
def divide_size(x):
  if x <= 40:
    return 1
  elif x <= 60:
    return 2
  elif x <= 85:
    return 3
  elif x <= 102:
    return 4
  elif x <= 135:
    return 5
  else:
    return 6
```

```python
data['면적기준'] = data['전용면적(㎡)'].apply(divide_size)
```

```python
data
```

**기준**


1.   40$m^2$ 이하
2.   40$m^2$ 초과 ~ 60$m^2$ 이하
3.   60$m^2$ 초과 ~ 85$m^2$ 이하
4.   85$m^2$ 초과 ~ 102$m^2$ 이하
5.   102$m^2$ 초과 ~ 135$m^2$ 이하
6.   135$m^2$ 초과



```python
def divide_size(x):
  if x <= 40:
    return 1
  elif x <= 60:
    return 2
  elif x <= 85:
    return 3
  elif x <= 102:
    return 4
  elif x <= 135:
    return 5
  else:
    return 6
```
