## 기존 CAPM 식을 그대로 사용해서 해보기

1. CAPM과 SMB 사용하기

```python
# X 데이터
X1 = pd.DataFrame(np.array(E_Rm.loc[:,['서울']]) - np.array(R_f), columns=['X1'])
SMB = SMB.reset_index(drop=True)
HML = HML.reset_index(drop=True)
# Y 데이터
Y = E_Ri.transpose() - R_f['국고채(3년,평균)']
Y = Y.transpose()

# train_set과 test_set 나누기 (7:3)
training_X1 = X1[:84]
training_SMB = SMB[:84]
training_HML = HML[:84]
training_Y = Y[:84]
test_X1 = X1[84:]
test_SMB = SMB[84:]
test_HML = HML[84:]
test_Y = Y[84:]
```

```python
score_list_1 = []
for i in range(training_Y.shape[1]):
  lr = LinearRegression(fit_intercept=True) # 절편항 있는 회귀 모형
  train_Xi = pd.concat([training_X1, training_SMB.iloc[:,i]], axis=1) # SMB 등은 여기서 추가하면 됨.
  train_Yi = training_Y.iloc[:,[i]] 
  lr.fit(train_Xi, train_Yi) 
  score_list_1 += [lr.score(train_Xi,train_Yi)]
```

```python
max(score_list_1)
```
